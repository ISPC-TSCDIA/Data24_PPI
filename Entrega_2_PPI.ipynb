#### Datos Ausentes

Lo primero que vamos a abordar es el tratamiento de datos ausentes, lo cual se presenta cuando hay celdas vacas. En este caso las soluciones pueden ser de dos tipos:
- rellenar la celda vaca con alguna de las siguientes opciones (entre muchas):

Esta solucin se utiliza cuando se dispone de grandes volmenes de datos, por tanto la eliminacin de alguna celda vaca no comprometer la cantidad o la calidad de la informacin disponible. Es la opcin ms sencilla y la que utilizamos a lo ltimo, para tener la oportunidad de generar alguna transformacin antes

#### Datos en Formato Incorrecto

Este quizs es el punto ms complicado que tengamos al momento de darle tratamiento al dataset con el que trabajamos, ya que va implicar pensar una estrategia para dejar los datos limpios.

La parte ms importante es la que se refiere a la columna `Motor`. Lo primero que haremos es comprobar el tipo de dato de algunas celdas especficas para averiguar con qu estamos trabajando, usando una combinacin de `loc` para acceder al valor de una celda, y `type` para obtener el tipo de dato:


````python
#type(data.loc[1, 'Motor'])
#type(data.loc[2, 'Motor'])
type(data.loc[10, 'Motor'])
#data['Motor'].dtypes
````

<br>

Haciendo diferentes pruebas vemos que la columna `Motor` es del tipo de dato `object`, que en nuestro caso se utiliza para representar cadenas de texto.

Lo que hacemos entonces es pasar todo directamente a `str` para poder manipular el contenido de las celdas, y luego pasar nuevamente todo a formato de `float` que es el que nos interesa:


````python
data['Motor'] = data['Motor'].astype(str)
````

<br>

Lo que vamos a hacer es buscar dentro de la columna `Motor` todas las celdas donde haya un punto, y modificaremos la celda para dejar nicamente el carcter que est antes del punto, el punto, y el carcter siguiente.

Para lograr esto vamos a usar una combinacin de expresiones regulares y la funcin `.apply()` de Pandas:


````python
# Primero debemos importar el mdulo 're' para trabajar con expresiones regulares
import re

# Luego definimos la funcin que realizar la operacin
def extraer_alrededor_punto(texto):
    match = re.search(r"(.)\.(.)", texto)
    if match:
        return match.group(0)  # Obtenemos los tres caracteres (antes, punto, despus)
    else:
        return texto  # Devolvemos el texto original si no hay punto

data['Motor'] = data['Motor'].apply(extraer_alrededor_punto)
#print(data['Motor'].to_string())
````

<br>

Ahora tenemos un problema: en algunas celdas se encuentra un valor de tipo `char` con un dgito, pero sin el `.0`, por lo tanto necesitamos una funcin que nos "empareje" esos datos.

Para ello volvemos a usar la funcin `.apply()` y generamos una funcin que toma un texto como entrada y con `.isdigit()` verifica si el texto contiene solo dgitos.
Si es as, devuelve el texto original con `.0` agregado al final, y si no, devuelve el texto original sin modificaciones.


````python
# Funcin para emparejar los datos
def agregar_punto_cero(texto):
    if texto.isdigit():  # Verificamos si el texto contiene solo dgitos
        return texto + '.0'  # Agregamos '.0' si es un dgito aislado
    else:
        return texto  # Devolvemos el texto original si no es un dgito aislado

data['Motor'] = data['Motor'].apply(agregar_punto_cero)
#print(data['Motor'].to_string())
#type(data.loc[13, 'Motor'])
````

<br>

Pero hasta aqu seguimos teniendo los valores dentro de las celdas de la columna `Motor` como cadenas de caracteres.

Para pasar todo a `float` primero necesitamos extraer slo los nmeros y luego convertirlos a punto flotante, otra vez utilizando las expresiones regulares. Por ltimo con `astype(float, errors='ignore')` intentamos convertir la columna a `float`. El parmetro `errors='ignore'` hace que se ignoren los valores que no se pueden convertir (como los `None` que pueden haber resultado de la extraccin):


````python
# Extraemos los nmeros (incluyendo el punto decimal)
def extraer_numeros(texto):
    match = re.search(r"\d+\.\d+", texto)  # Buscamos uno o ms dgitos, punto, uno o ms dgitos
    if match:
        return match.group(0)
    else:
        return None  # Devolvemos None si no hay nmeros

data['Motor'] = data['Motor'].apply(extraer_numeros)

# Convertimos a float (ignorando los valores no numricos)
data['Motor'] = data['Motor'].astype(float, errors='ignore')

#print(data['Motor'].to_string())
````

````python
#type(data.loc[13, 'Motor'])
data['Motor'].dtypes
````

_Como se puede observar, las operaciones han sido exitosas: tenemos la columna Motor slo con valores numricos de punto flotante._

<br>

Lo que haremos es sencillo, comprobaremos si algn valor es mayor que "3000" y procederemos entonces a quitarle ceros hasta dejarlo de cuatro dgitos. Esto lo vamos a lograr tomando un valor de ao como entrada y si el ao es mayor que 3000, entramos en un bucle `while` que se ejecuta mientras el ao sea mayor que 9999 (es decir, tenga ms de cuatro dgitos). Dentro del bucle, con `ao //= 10` hacemos una divisin entera entre 10, eliminando el ltimo dgito del ao, hasta que finalmente obtenemos el ao corregido.:


````python
def corregir_ao(ao):
    if ao > 3000:
        while ao > 9999:
            ao //= 10  # Dividimos el ao entre 10 (elimina el ltimo dgito)
    return ao

data['Ao'] = data['Ao'].apply(corregir_ao)
#print(data['Ao'].to_string())
````

### Identificacin de Outliers

Una parte importante de un anlisis univariado es detectar posibles outliers en las variables numricas. Para ello se pueden utilizar tcnicas como el anlisis de boxplots o el clculo del z-score.

En nuestro caso los valores atpicos o extremos fueron tratados durante el curado de los datos, pero a efectos de asegurarnos que no existan aun, o simplemente por prctica/aprendizaje, se proceder a realizar un clculo del z-score de la variable/columna `Ao`.

Para ello primero se debe importar el mdulo `stats` de la librera `scipy`.


````python
from scipy import stats
````

Luego, se calcula el z-score de la columna `Ao` usando `stats.zscore()` y se lo guarda en una nueva columna llamada `Ao_zscore`:

````python
data['Ao_zscore'] = stats.zscore(data['Ao'])
````

Ahora bien, el z-score ayuda a determinar si existe un valor atpico al indicar cun lejos se encuentra un punto de datos de la media en trminos de desviaciones estndar.

(Generalmente, un valor atpico se define como un punto de datos con un z-score mayor que 3 o menor que -3. Esto se debe a que en una distribucin normal, aproximadamente el 99.7% de los datos se encuentran dentro de 3 desviaciones estndar de la media.)

Por lo tanto, si el z-score de un dato en la columna 'Precio' es mayor que 3 o menor que -3, puede ser considerado un valor atpico.

Como sera imprctico revisar todos los datos de una columna individualmente, lo que se hace entonces es buscar dentro de la columna `Ao_zscore` si aparece algn valor mayor que 3 o menor que -3, y determinar dnde se encuentra:

````python
# Se encuentran las filas donde los valores en la columna 'Ao' son mayores
# que 3 o menores que -3.
filas = data[(data['Ao_zscore'] > 3) | (data['Ao_zscore'] < -3)]

# Se imprimen las filas o un mensaje si no se encuentra ninguna.
if filas.empty:
  print("No se encontraron valores mayores que 3 o menores que -3.")
else:
  print("Filas con valores mayores que 3 o menores que -3:")
  print(filas.index)
````

_Para el dataset con el que se est trabajando, se puede observar que existen cuatro entradas de datos con un valor de `Ao` fuera de las tres desviaciones estndar de la media, pero al constatar estos datos individualmente se observa que no es un problema._
